```python
import requests
import csv
import pandas as pd

response = requests.get("https://gorest.co.in/public/v2/posts")
result = response.json()
result
```

```python
result[0]
```

```python
with open("employees.csv",'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=['id', 'user_id', 'title', 'body'])
    writer.writeheader()
    for employees in result:
        i_d = employees['id']
        user_id = employees['user_id']
        title = employees['title']
        body = employees['body']
        writer.writerow(employees)
df = pd.read_csv('employees.csv')
df
```

```python
result_users = requests.get("https://gorest.co.in/public/v2/users")
result = result_users.json()
result[0]
```

```python
with open("status.csv",'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=['id', 'name', 'email', 'gender','status'])
    writer.writeheader()
    for employees in result:
        i_d = employees['id']
        name = employees['name']
        email = employees['email']
        gender = employees['gender']
        status = employees['status']
        writer.writerow(employees)
df = pd.read_csv('status.csv')
df
```
